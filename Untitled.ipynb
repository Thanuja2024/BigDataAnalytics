## Dataset Insights

The dataset provided contains details of 1,000 publicly listed companies across various sectors, industries, and countries. It consists of seven columns — symbol, name, sector, industry, marketCap, country, and ipoYear. The data captures company-level information such as industry classification, market capitalization, and IPO year, enabling comprehensive analysis of business and financial trends. The dataset is complete and free from missing values, ensuring data consistency and analytical reliability.

**1. Structure and Uniqueness**

The dataset comprises 1,000 rows and 7 columns, with each company uniquely identified by its stock symbol. The name field contains 1,000 unique company names, confirming no duplication of entities. The dataset spans six major sectors and multiple industries, offering a broad view of the market landscape. Companies originate from five different countries, providing geographical diversity suitable for comparative regional and sectoral analysis.

**2. Sectoral Distribution**

There are six sectors represented in the dataset, with the Technology sector having the highest number of companies, followed by Healthcare, Finance, and Consumer Services. This distribution reflects the global market trend where technology-driven organizations dominate modern stock markets. The presence of diverse sectors ensures balanced analysis across industries, from service-based firms to capital-intensive enterprises.

**3. Country-Wise Representation**

The dataset covers companies from five countries, with the United States (US) having the largest representation, followed by Canada, United Kingdom (UK), Germany, and Japan. The dominance of US-based companies aligns with the country’s strong financial markets and concentration of tech giants. This regional spread highlights global participation, suitable for exploring how economic zones influence company growth and sector performance.

**4. Market Capitalization Insights**

The marketCap variable demonstrates a wide range, from a few million to several hundred billion dollars. The average market capitalization is approximately 50 billion, with a standard deviation indicating significant variation between small-cap and large-cap firms. The median market cap sits around 45 billion, suggesting a moderately right-skewed distribution due to a few extremely high-valued companies (outliers), such as multinational technology firms. This variation allows for meaningful insights into company scale and investment magnitude across industries.

**5. IPO Year Distribution**

The dataset includes companies with IPO years ranging from 1950 to 2024, reflecting both historical and modern market entries. Most companies went public between 1990 and 2015, indicating the globalization and technological expansion eras. The IPO trend analysis can help understand industry emergence patterns, market cycles, and economic phases influencing company listings.

**6. Industry Spread**

Within each sector, multiple industries are represented — for instance, Technology includes “Software,” “Semiconductors,” and “IT Services,” while Healthcare covers “Pharmaceuticals” and “Biotechnology.” The diversity across industries provides a rich base for segmentation and performance analysis, revealing which areas dominate or are underrepresented in the dataset.

**7. Key Relationships**

Sector vs. MarketCap: Sectors such as Technology and Finance tend to have higher market capitalizations, reflecting investor confidence and innovation-driven growth.

Country vs. Sector: Some sectors are more concentrated in specific countries — e.g., Technology in the US, Automotive in Germany, and Finance in the UK.

IPO Year vs. MarketCap: More recently listed companies may exhibit higher market caps, especially in tech-driven sectors, showing the evolving dominance of digital-era firms.

**8. Data Characteristics**

The dataset provides a comprehensive snapshot of global company data, ideal for financial trend analysis, market segmentation, sector growth evaluation, and investment potential studies. The absence of missing values ensures high-quality analytics and accurate modeling. It serves as an excellent foundation for Big Data Analytics tasks such as:

Predicting company growth patterns

Sector-wise profitability comparison

Visualizing IPO trends and geographic distributions

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, count, mean, max, min

# Initialize SparkSession
spark = SparkSession.builder \
    .appName("NASDAQ Analysis") \
    .getOrCreate()

# Load CSV
df = spark.read.csv("symbols_valid_meta.csv", header=True, inferSchema=True)

# Show schema
df.printSchema()

# Show sample data
df.show(5)
```

```python
total_rows = df.count()
print("Total rows:", total_rows)
```

```python
distinct_symbols = df.select("Symbol").distinct().count()
print("Distinct symbols:", distinct_symbols)
```

```python
df.groupBy("ETF").count().show()
```

```python
df.groupBy("Financial Status").count().show()
```

```python
df.select(max("Round Lot Size"), min("Round Lot Size")).show()
```

```python
df.select(mean("Round Lot Size")).show()
```

```python
df.filter(col("Nasdaq Traded") == "Y").show(5)
```

```python
df.groupBy("Listing Exchange").count().show()
```

```python
df.orderBy(col("Round Lot Size").desc()).show(10)
```

```python
import matplotlib.pyplot as plt

# Convert grouped results to pandas
etf_counts = df.groupBy("ETF").count().toPandas()
exchange_counts = df.groupBy("Listing Exchange").count().toPandas()
financial_counts = df.groupBy("Financial Status").count().toPandas()
market_counts = df.groupBy("Market Category").count().toPandas()
```

```python
etf_counts.plot.pie(y='count', labels=etf_counts['ETF'], autopct='%1.1f%%', legend=False)
plt.title("ETF vs Non-ETF")
plt.show()
```

```python
exchange_counts.plot.bar(x='Listing Exchange', y='count', color='skyblue')
plt.title("Securities by Listing Exchange")
plt.show()
```

```python
financial_counts.plot.bar(x='Financial Status', y='count', color='orange')
plt.title("Securities by Financial Status")
plt.show()
```

```python
market_counts.plot.bar(x='Market Category', y='count', color='green')
plt.title("Market Category Distribution")
plt.show()
```

```python
df.select("Round Lot Size").toPandas().plot.hist(bins=20, color='purple')
plt.title("Round Lot Size Distribution")
plt.show()
```

```python
nasdaq_counts = df.groupBy("Nasdaq Traded").count().toPandas()
nasdaq_counts.plot.bar(x='Nasdaq Traded', y='count', color='red')
plt.title("NASDAQ Traded vs Non-Traded")
plt.show()
```

```python
top10_roundlot = df.orderBy(col("Round Lot Size").desc()).limit(10).toPandas()
top10_roundlot.plot.bar(x='Symbol', y='Round Lot Size', color='blue')
plt.title("Top 10 Securities by Round Lot Size")
plt.show()
```

```python
nextshares_counts = df.groupBy("NextShares").count().toPandas()
nextshares_counts.plot.bar(x='NextShares', y='count', color='cyan')
plt.title("NextShares vs Non-NextShares")
plt.show()
```

```python
market_etf = df.groupBy("Market Category", "ETF").count().toPandas()
market_etf.pivot(index='Market Category', columns='ETF', values='count').plot.bar(stacked=True)
plt.title("Market Category vs ETF Count")
plt.show()
```
